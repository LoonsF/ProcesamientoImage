```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def solarizar(imagen, umbral):
    imagen_solarizada = np.where(imagen < umbral, imagen, 255 - imagen)
    return imagen_solarizada

imagen = cv2.imread('image.jpg')
img_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

umbral = 127
imagen_nueva = solarizar(img_gris, umbral)

imagen_falso_color = cv2.applyColorMap(imagen_nueva, cv2.COLORMAP_JET)

plt.figure(figsize=(10, 8))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title("Imagen Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(imagen_nueva, cmap='gray')
plt.title("Solarizada en Gris")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(imagen_falso_color, cv2.COLOR_BGR2RGB))
plt.title("Falso Color")
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def falso_color(imagen_gray):
    escalar_imagen = cv2.normalize(imagen_gray, None, 0, 255, cv2.NORM_MINMAX)
    aplica_color = cv2.applyColorMap(escalar_imagen, cv2.COLORMAP_INFERNO)
    return aplica_color

img_gray = cv2.imread('image.jpg', 0)  
img_falso = falso_color(img_gray)

plt.figure(figsize=(10, 8))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img_falso, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

paletas = {
    cv2.COLORMAP_AUTUMN: "AUTUMN",
    cv2.COLORMAP_BONE: "BONE",
    cv2.COLORMAP_JET: "JET",
    cv2.COLORMAP_WINTER: "WINTER",
    cv2.COLORMAP_RAINBOW: "RAINBOW",
    cv2.COLORMAP_OCEAN: "OCEAN",
    cv2.COLORMAP_SUMMER: "SUMMER",
    cv2.COLORMAP_SPRING: "SPRING",
    cv2.COLORMAP_COOL: "COOL",
    cv2.COLORMAP_HSV: "HSV",
    cv2.COLORMAP_HOT: "HOT",
    cv2.COLORMAP_PINK: "PINK",
    cv2.COLORMAP_PARULA: "PARULA",
    cv2.COLORMAP_MAGMA: "MAGMA",
    cv2.COLORMAP_INFERNO: "INFERNO",
    cv2.COLORMAP_PLASMA: "PLASMA",
    cv2.COLORMAP_VIRIDIS: "VIRIDIS",
    cv2.COLORMAP_CIVIDIS: "CIVIDIS",
    cv2.COLORMAP_TWILIGHT: "TWILIGHT",
    cv2.COLORMAP_TWILIGHT_SHIFTED: "TWILIGHT_SHIFTED",
    cv2.COLORMAP_TURBO: "TURBO",
    cv2.COLORMAP_DEEPGREEN: "DEEPGREED"
}

image_path = "image.jpg"
img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(24, 15))
for i, nombre in enumerate(paletas):
    img_color = cv2.applyColorMap(img_gray, nombre)
    
    plt.subplot(6, 4, i + 1)
    plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
    plt.title(paletas[nombre])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
